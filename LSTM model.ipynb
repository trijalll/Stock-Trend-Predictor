```python
#splitting data into training and testing 

data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])#we will be starting from 0th index to 70% of the total values
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])#for testing it will be starting from 70% to len of df

print(data_training.shape)
print(data_testing.shape)
```

```python
#training data
data_training.head()
```

```python
#testing data
data_testing.head()
```

```python
#scalling down the data b/w 0 and 1 for that we use min max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array#stored training data into array
```

```python
#divide data in y train and xtrain
x_train=[]#values of previous 100 days though which we will predict value on 101st day
y_train=[]#predicted value

for i in range(100,data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i,0])
    
x_train,y_train = np.array(x_train),np.array(y_train)
```

```python
#ML MODEL
```

```python
from keras.layers import Dense,Dropout,LSTM
from keras.models import Sequential
```

```python
#creating the model
model = Sequential()
#1st layer
model.add(LSTM(units = 50,activation = 'relu',return_sequences = True,
              input_shape = (x_train.shape[1],1)))
model.add(Dropout(0.2))

#2nd layer
model.add(LSTM(units = 60,activation = 'relu',return_sequences = True))
model.add(Dropout(0.3))

#3rd layer
model.add(LSTM(units = 80,activation = 'relu',return_sequences = True))
model.add(Dropout(0.4))

#4th layer
model.add(LSTM(units = 120,activation = 'relu'))
model.add(Dropout(0.5))

#last layer to connect all these layers AKA the dense layer
model.add(Dense(units=1))
```

```python
#summary of the model
model.summary()
```

```python
#compiling the model
model.compile(optimizer='adam', loss = 'mean_squared_error')
model.fit(x_train,y_train,epochs = 50)
```

```python
#now the model is trained so now we are saving this model so that 
#we can use this model in our stream lit web app 
model.save('keras_model.h5')
```
